```python
from collections import Counter
```

```python
data=[1,1,1,2,2,23,3,33,13,12,23,34,33,5,5,5,5]
```

```python
absolute_freq=Counter(data)
for k,v in absolute_freq.items():
    print(f"{k} : {v}")
```

```python
total=sum(absolute_freq.values())
total
```

```python
relative_freq={k:v/total for k,v in absolute_freq.items()}
for k,v in relative_freq.items():
     print(f"relative frequency of {k} is  {v}")
```

```python
cum=0
for k,v in sorted(absolute_freq.items()):
    cum+=v
    print(f"cumulative frequency of {k} is {cum}")
```
