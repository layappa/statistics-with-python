# using pandas

```python
import pandas as pd

# Sample data
data = [10, 20, 30, 20, 40, 50, 20]

# Convert to pandas Series
s = pd.Series(data)

# Mean
mean_value = s.mean()

# Median
median_value = s.median()

# Mode (Pandas may return multiple modes)
mode_value = s.mode()

print("Mean :", mean_value)
print("Median :", median_value)
print("Mode :", list(mode_value))
```

# using numpy

```python
import numpy as np
from statistics import mode, multimode

# Sample data
data = [10, 20, 30, 20, 40, 50, 20]

arr = np.array(data)

# Mean
mean_value = np.mean(arr)

# Median
median_value = np.median(arr)

# Mode (statistics module for mode)
mode_value = mode(arr)       # single mode
all_modes = multimode(arr)   # returns list if multiple

print("Mean :", mean_value)
print("Median :", median_value)
print("Mode :", mode_value)
print("All Modes :", all_modes)
```
